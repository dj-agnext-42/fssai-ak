```python
import pandas as pd
from joblib import Parallel, delayed

# Read both sheets from the Excel file
df_sheet1 = pd.read_excel('param_updated_with_limits.xlsx')
df_master = df_new_Master

# Create a dictionary for quick lookup from df_master
master_dict = df_master.set_index(['Commodity', 'Parameter'])[
    ['Max FSSAI Acceptable Value', 'Max FSSAI Acceptable Value - Organic']
].to_dict()

# Function to get the appropriate value
def get_master_value(commodity, parameter, variant):
    try:
        if (commodity, parameter) in master_dict['Max FSSAI Acceptable Value']:
            if variant in ['Loose', 'Normal']:
                return master_dict['Max FSSAI Acceptable Value'][(commodity, parameter)]
            elif variant == 'Organic':
                return master_dict['Max FSSAI Acceptable Value - Organic'][(commodity, parameter)]
    except KeyError:
        return None

# Identify parameter columns dynamically
param_columns = [col for col in df_sheet1.columns if col not in ['Commodity', 'Variant'] and not col.startswith('FSSR Limits')]

# Function to process a single row
def process_row(row):
    for param in param_columns:
        fssr_col = f'FSSR Limits {param}'
        master_value = get_master_value(row['Commodity'], param, row['Variant'])
        
        if master_value is not None:
            row[fssr_col] = f"{row[fssr_col]} {master_value}"
    return row

# Apply parallel processing using joblib
processed_rows = Parallel(n_jobs=-1)(delayed(process_row)(row) for _, row in df_sheet1.iterrows())

# Convert list of dictionaries back to DataFrame
df_sheet1 = pd.DataFrame(processed_rows)

# Save the updated DataFrame
df_sheet1.to_excel('b.xlsx', index=False)
print("Successfully updated all FSSR Limits columns with values from Master sheet")
```

```python
# import pandas as pd
# from joblib import Parallel, delayed

# # Read the file with current FSSR limits and values
# df_sheet1 = pd.read_excel('param_updated_with_all_limits_and_values.xlsx')
# df_master = df_new_Master

# # Create a dictionary for quick lookup from df_master
# unit_dict = df_master.set_index(['Commodity', 'Parameter'])['Units'].to_dict()

# # Function to get unit from master sheet
# def get_unit_from_master(commodity, parameter):
#     return unit_dict.get((commodity, parameter), None)

# # Get all FSSR Limits columns
# fssr_columns = [col for col in df_sheet1.columns if col.startswith('FSSR Limits')]

# # Function to process a single row
# def process_row(row):
#     commodity = row['Commodity']
#     for fssr_col in fssr_columns:
#         param = fssr_col.replace('FSSR Limits ', '')
#         unit = get_unit_from_master(commodity, param)
        
#         if unit is not None:
#             current_value = str(row[fssr_col])
#             if unit not in current_value:
#                 row[fssr_col] = f"{current_value} {unit}"
#     return row

# # Apply parallel processing using joblib
# processed_rows = Parallel(n_jobs=-1)(delayed(process_row)(row) for _, row in df_sheet1.iterrows())

# # Convert list of dictionaries back to DataFrame
# df_sheet1 = pd.DataFrame(processed_rows)

# # Save the updated DataFrame
# df_sheet1.to_excel('param_updated_with_limits_values_and_units.xlsx', index=False)
# print("Successfully added units to FSSR Limits columns")
# ************************************************************************************
# # Read both sheets from the Excel file
# import pandas as pd

# df_sheet1 = pd.read_excel('b.xlsx')
# df_master = df_new_Master

# # Identify the relevant parameter columns based on the given range
# start_col = 'Acetamiprid, mg/kg'
# end_col = 'FSSR Limits Copper Oxychloride (Copper determined as elemental copper), mg/ kg'

# # Get the list of columns that fall within the specified range
# param_columns = list(df_sheet1.columns)
# start_index = param_columns.index(start_col)
# end_index = param_columns.index(end_col)
# selected_params = param_columns[start_index:end_index+1]  # Ensure inclusive range

# # Extract only parameter names (exclude 'FSSR Limits' ones)
# params = [col for col in selected_params if not col.startswith('FSSR Limits')]

# # Function to get the appropriate value from Master sheet
# def get_master_value(commodity, parameter, variant):
#     master_row = df_master[(df_master['Commodity'] == commodity) & 
#                           (df_master['Parameter'] == parameter)]
    
#     if not master_row.empty:
#         if variant in ['Loose', 'Normal']:
#             value = master_row['Max FSSAI Acceptable Value'].iloc[0]
#         elif variant == 'Organic':
#             value = master_row['Max FSSAI Acceptable Value - Organic'].iloc[0]
#         else:
#             return None
        
#         if pd.isna(value):
#             return None
#         return str(value)
#     return None

# # Iterate through each row in Sheet1
# for idx, row in df_sheet1.iterrows():
#     commodity = row['Commodity']
#     variant = row['Variant']
    
#     for param in params:
#         fssr_col = f'FSSR Limits {param}'
#         compliance_col = f'{param} Compliance'
        
#         # Ensure the compliance column exists
#         if compliance_col not in df_sheet1.columns:
#             df_sheet1[compliance_col] = None
        
#         current_limit = df_sheet1.at[idx, fssr_col] if fssr_col in df_sheet1.columns else None
#         current_value = df_sheet1.at[idx, param] if param in df_sheet1.columns else None
        
#         # Always use 'Max' as limit type
#         master_value = get_master_value(commodity, param, variant)
        
#         if master_value is not None:
#             # Update the FSSR Limits column with Max value
#             df_sheet1.at[idx, fssr_col] = f"{current_limit} {master_value}"
            
#             # Extract the numeric part of FSSR Limits
#             try:
#                 limit_value = float(str(master_value).split()[0])  # Ensure numeric conversion
#             except ValueError:
#                 limit_value = None
            
#             if limit_value is not None and pd.notna(current_value):
#                 # Get decimal places from limit value
#                 decimal_places = len(str(limit_value).split('.')[-1]) if '.' in str(limit_value) else 0
                
#                 # Round the current value to the same decimal places
#                 rounded_value = round(float(current_value), decimal_places)
                
#                 # Check compliance
#                 df_sheet1.at[idx, compliance_col] = 'Safe' if rounded_value <= limit_value else 'Unsafe'

# # Save the updated DataFrame
# df_sheet1.to_excel('c.xlsx', index=False)
# print("Successfully updated all FSSR Limits columns and added compliance checks")

# Read both sheets from the Excel file
# import pandas as pd

# df_sheet1 = pd.read_excel('b.xlsx')
# df_master = df_new_Master

# # Identify the relevant parameter columns based on the given range
# start_col = 'Acetamiprid, mg/kg'
# end_col = 'FSSR Limits Copper Oxychloride (Copper determined as elemental copper), mg/ kg'

# # Get the list of columns that fall within the specified range
# param_columns = list(df_sheet1.columns)
# start_index = param_columns.index(start_col)
# end_index = param_columns.index(end_col)
# selected_params = param_columns[start_index:end_index+1]  # Ensure inclusive range

# # Extract only parameter names (exclude 'FSSR Limits' ones)
# params = [col for col in selected_params if not col.startswith('FSSR Limits')]

# # Function to get the appropriate value from Master sheet
# def get_master_value(commodity, parameter, variant):
#     master_row = df_master[(df_master['Commodity'] == commodity) & 
#                           (df_master['Parameter'] == parameter)]
    
#     if not master_row.empty:
#         if variant in ['Loose', 'Normal']:
#             value = master_row['Max FSSAI Acceptable Value'].iloc[0]
#         elif variant == 'Organic':
#             value = master_row['Max FSSAI Acceptable Value - Organic'].iloc[0]
#         else:
#             return None
        
#         if pd.isna(value):
#             return None
#         return str(value)
#     return None

# # Iterate through each row in Sheet1
# for idx, row in df_sheet1.iterrows():
#     commodity = row['Commodity']
#     variant = row['Variant']
    
#     for param in params:
#         fssr_col = f'FSSR Limits {param}'
#         compliance_col = f'{param} Compliance'
        
#         # Ensure the compliance column exists
#         if compliance_col not in df_sheet1.columns:
#             df_sheet1[compliance_col] = None
        
#         current_limit = df_sheet1.at[idx, fssr_col] if fssr_col in df_sheet1.columns else None
#         current_value = df_sheet1.at[idx, param] if param in df_sheet1.columns else None
        
#         # If current_value is missing, leave compliance blank and continue
#         if pd.isna(current_value) or current_value == '':
#             df_sheet1.at[idx, compliance_col] = ''
#             continue  # Skip further processing for this parameter
        
#         # Always use 'Max' as limit type
#         master_value = get_master_value(commodity, param, variant)
        
#         if master_value is not None:
#             # Update the FSSR Limits column with Max value
#             df_sheet1.at[idx, fssr_col] = f"{current_limit} {master_value}"
            
#             # Extract the numeric part of FSSR Limits
#             try:
#                 limit_value = float(str(master_value).split()[0])  # Ensure numeric conversion
#             except ValueError:
#                 limit_value = None
            
#             if limit_value is not None:
#                 # Get decimal places from limit value
#                 decimal_places = len(str(limit_value).split('.')[-1]) if '.' in str(limit_value) else 0
                
#                 # Round the current value to the same decimal places
#                 rounded_value = round(float(current_value), decimal_places)
                
#                 # Check compliance
#                 df_sheet1.at[idx, compliance_col] = 'Safe' if rounded_value <= limit_value else 'Unsafe'

# # Save the updated DataFrame
# df_sheet1.to_excel('c.xlsx', index=False)
# print("Successfully updated all FSSR Limits columns and added compliance checks")
# Read both sheets from the Excel file
# import pandas as pd

# df_sheet1 = pd.read_excel('b.xlsx')
# df_master = df_new_Master

# # Identify the relevant parameter columns based on the given range
# start_col = 'Acetamiprid, mg/kg'
# end_col = 'FSSR Limits Copper Oxychloride (Copper determined as elemental copper), mg/ kg'

# # Get the list of columns that fall within the specified range
# param_columns = list(df_sheet1.columns)
# start_index = param_columns.index(start_col)
# end_index = param_columns.index(end_col)
# selected_params = param_columns[start_index:end_index+1]  # Ensure inclusive range

# # Extract only parameter names (exclude 'FSSR Limits' ones)
# params = [col for col in selected_params if not col.startswith('FSSR Limits')]

# # Function to get the appropriate value from Master sheet
# def get_master_value(commodity, parameter, variant):
#     master_row = df_master[(df_master['Commodity'] == commodity) & 
#                           (df_master['Parameter'] == parameter)]
    
#     if not master_row.empty:
#         if variant in ['Loose', 'Normal']:
#             value = master_row['Max FSSAI Acceptable Value'].iloc[0]
#         elif variant == 'Organic':
#             value = master_row['Max FSSAI Acceptable Value - Organic'].iloc[0]
#         else:
#             return None
        
#         if pd.isna(value):
#             return None
#         return str(value)
#     return None

# # Iterate through each row in Sheet1
# for idx, row in df_sheet1.iterrows():
#     commodity = row['Commodity']
#     variant = row['Variant']
    
#     for param in params:
#         fssr_col = f'FSSR Limits {param}'
#         compliance_col = f'{param} Compliance'
        
#         if compliance_col not in df_sheet1.columns:
#             df_sheet1[compliance_col] = None

#         current_limit = df_sheet1.at[idx, fssr_col] if fssr_col in df_sheet1.columns else None
#         current_value = df_sheet1.at[idx, param] if param in df_sheet1.columns else None
        
#         # Always get the master value for FSSR Limits
#         master_value = get_master_value(commodity, param, variant)

#         if master_value is not None:
#             df_sheet1.at[idx, fssr_col] = f"{current_limit} {master_value}"

#         # If the current_value is empty, skip compliance check but keep FSSR Limits updated
#         if pd.isna(current_value) or current_value == '':
#             df_sheet1.at[idx, compliance_col] = ''
#             continue

#         # Extract numeric limit value
#         try:
#             limit_value = float(str(master_value).split()[0])
#         except ValueError:
#             limit_value = None

#         if limit_value is not None:
#             decimal_places = len(str(limit_value).split('.')[-1]) if '.' in str(limit_value) else 0
#             rounded_value = round(float(current_value), decimal_places)
#             df_sheet1.at[idx, compliance_col] = 'Safe' if rounded_value <= limit_value else 'Unsafe'


# # Save the updated DataFrame
# df_sheet1.to_excel('c.xlsx', index=False)
# print("Successfully updated all FSSR Limits columns and added compliance checks")
# import pandas as pd

# df_sheet1 = pd.read_excel('b.xlsx')
# df_master = df_new_Master

# # Identify the relevant parameter columns based on the given range
# start_col = 'Acetamiprid, mg/kg'
# end_col = 'FSSR Limits Copper Oxychloride (Copper determined as elemental copper), mg/ kg'

# # Get the list of columns that fall within the specified range
# param_columns = list(df_sheet1.columns)
# start_index = param_columns.index(start_col)
# end_index = param_columns.index(end_col)
# selected_params = param_columns[start_index:end_index+1]  # Ensure inclusive range

# # Extract only parameter names (exclude 'FSSR Limits' ones)
# params = [col for col in selected_params if not col.startswith('FSSR Limits')]

# # Function to get the appropriate value from Master sheet
# def get_master_value(commodity, parameter, variant):
#     master_row = df_master[(df_master['Commodity'] == commodity) & 
#                            (df_master['Parameter'] == parameter)]
    
#     if not master_row.empty:
#         if variant in ['Loose', 'Normal']:
#             value = master_row['Max FSSAI Acceptable Value'].iloc[0]
#         elif variant == 'Organic':
#             value = master_row['Max FSSAI Acceptable Value - Organic'].iloc[0]
#         else:
#             return None
        
#         if pd.isna(value):
#             return None
#         return str(value)
#     return None

# # Iterate through each row in Sheet1
# for idx, row in df_sheet1.iterrows():
#     commodity = row['Commodity']
#     variant = row['Variant']
    
#     for param in params:
#         fssr_col = f'FSSR Limits {param}'
#         compliance_col = f'{param} Compliance'
        
#         # Ensure the compliance column exists
#         if compliance_col not in df_sheet1.columns:
#             df_sheet1[compliance_col] = None

#         current_limit = df_sheet1.at[idx, fssr_col] if fssr_col in df_sheet1.columns else None
#         current_value = df_sheet1.at[idx, param] if param in df_sheet1.columns else None
        
#         # Always get the master value for FSSR Limits
#         master_value = get_master_value(commodity, param, variant)

#         if master_value is not None:
#             df_sheet1.at[idx, fssr_col] = f"{current_limit} {master_value}"

#         # If the current_value is empty, skip the compliance check
#         if pd.isna(current_value) or current_value == '':
#             df_sheet1.at[idx, compliance_col] = ''
#             continue

#         # Extract the numeric part of master_value as a string
#         master_value_str = str(master_value).split()[0]
#         try:
#             limit_value = float(master_value_str)
#         except ValueError:
#             limit_value = None

#         if limit_value is not None:
#             # Determine decimal places from the original string to maintain precision
#             if '.' in master_value_str:
#                 decimal_places = len(master_value_str.split('.')[1])
#             else:
#                 decimal_places = 0

#             # Round the current value to the same number of decimal places
#             try:
#                 rounded_value = round(float(current_value), decimal_places)
#             except ValueError:
#                 rounded_value = None

#             if rounded_value is not None:
#                 df_sheet1.at[idx, compliance_col] = 'Safe' if rounded_value <= limit_value else 'Unsafe'

# # Save the updated DataFrame
# df_sheet1.to_excel('c.xlsx', index=False)
# print("Successfully updated all FSSR Limits columns and added compliance checks")
# import pandas as pd
import pandas as pd
from decimal import Decimal, ROUND_HALF_UP
from joblib import Parallel, delayed

def round_half_up(value, decimals):
    rounding_quantizer = Decimal('1.' + '0' * decimals)
    return float(Decimal(value).quantize(rounding_quantizer, rounding=ROUND_HALF_UP))

df_sheet1 = pd.read_excel('b.xlsx')
df_master = df_new_Master

# Identify the relevant parameter columns based on the given range
start_col = 'Acetamiprid, mg/kg'
end_col = 'FSSR Limits Copper Oxychloride (Copper determined as elemental copper), mg/ kg'

# Get the list of columns that fall within the specified range
param_columns = list(df_sheet1.columns)
start_index = param_columns.index(start_col)
end_index = param_columns.index(end_col)
selected_params = param_columns[start_index:end_index+1]  # Ensure inclusive range

# Extract only parameter names (exclude 'FSSR Limits' ones)
params = [col for col in selected_params if not col.startswith('FSSR Limits')]

# Function to get the appropriate value from Master sheet
def get_master_value(commodity, parameter, variant):
    master_row = df_master[(df_master['Commodity'] == commodity) & 
                           (df_master['Parameter'] == parameter)]
    if not master_row.empty:
        if variant in ['Loose', 'Normal']:
            value = master_row['Max FSSAI Acceptable Value'].iloc[0]
        elif variant == 'Organic':
            value = master_row['Max FSSAI Acceptable Value - Organic'].iloc[0]
        else:
            return None
        if pd.isna(value):
            return None
        return str(value)
    return None

# Function to process a single row (optimized using parallel processing)
def process_row(row):
    row = row.copy()  # Ensure we are not modifying the original row
    commodity = row['Commodity']
    variant = row['Variant']
    
    for param in params:
        fssr_col = f'FSSR Limits {param}'
        compliance_col = f'{param} Compliance'
        
        # Ensure the compliance column exists in the row
        if compliance_col not in row:
            row[compliance_col] = None

        current_limit = row[fssr_col] if fssr_col in row else None
        current_value = row[param] if param in row else None
        
        # Always get the master value for FSSR Limits
        master_value = get_master_value(commodity, param, variant)
        if master_value is not None:
            row[fssr_col] = f"{current_limit} {master_value}"

        # If the current_value is empty, skip the compliance check
        if pd.isna(current_value) or current_value == '':
            row[compliance_col] = ''
            continue

        # Extract the numeric part of master_value as a string
        master_value_str = str(master_value).split()[0] if master_value is not None else ""
        try:
            limit_value = float(master_value_str)
        except ValueError:
            limit_value = None

        if limit_value is not None:
            # Determine decimal places from the original string to maintain precision
            if '.' in master_value_str:
                decimal_places = len(master_value_str.split('.')[1])
            else:
                decimal_places = 0

            # Round the current value to the same number of decimal places using round_half_up
            try:
                rounded_value = round_half_up(float(current_value), decimal_places)
            except ValueError:
                rounded_value = None

            if rounded_value is not None:
                row[compliance_col] = 'Safe' if rounded_value <= limit_value else 'Unsafe'
    return row

# Apply parallel processing to each row
processed_rows = Parallel(n_jobs=-1)(delayed(process_row)(row) for _, row in df_sheet1.iterrows())

# Convert list of processed rows back to DataFrame
df_sheet1 = pd.DataFrame(processed_rows)

# Save the updated DataFrame
df_sheet1.to_excel('c.xlsx', index=False)
print("Successfully updated all FSSR Limits columns and added compliance checks")
```
