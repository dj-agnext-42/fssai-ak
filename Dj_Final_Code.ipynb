```python
# Read both sheets from the Excel file
import pandas as pd

df_sheet1 = pd.read_excel('param_updated_with_limits.xlsx')
df_master = df_new_Master

# Function to get the appropriate value from Master sheet
def get_master_value(commodity, parameter, variant, limit_type='Max'):
    # Find matching row in master sheet
    master_row = df_master[(df_master['Commodity'] == commodity) & 
                          (df_master['Parameter'] == parameter)]
    
    if not master_row.empty:
        if variant in ['Loose', 'Normal']:
            value = master_row['Max FSSAI Acceptable Value'].iloc[0]
        elif variant == 'Organic':
            value = master_row['Max FSSAI Acceptable Value - Organic'].iloc[0]
        else:
            return None
        
        # Convert to string and handle NaN
        if pd.isna(value):
            return None
        return str(value)
    return None

# Iterate through each row in Sheet1
for idx, row in df_sheet1.iterrows():
    commodity = row['Commodity']
    variant = row['Variant']
    
    # Process all columns except 'Commodity' and 'Variant'
    param_columns = [col for col in df_sheet1.columns if col not in ['Commodity', 'Variant'] 
                    and not col.startswith('FSSR Limits')]
    
    for param in param_columns:
        fssr_col = f'FSSR Limits {param}'
        current_limit = df_sheet1.at[idx, fssr_col]
        
        # Always use 'Max' as limit type
        master_value = get_master_value(commodity, param, variant, 'Max')
        
        if master_value is not None:
            # Update the FSSR Limits column with Max value
            df_sheet1.at[idx, fssr_col] = f"{current_limit} {master_value}"

# Save the updated DataFrame
df_sheet1.to_excel('param_updated_with_all_limits_and_values.xlsx', index=False)
print("Successfully updated all FSSR Limits columns with values from Master sheet")
```

```python
# Read the file with current FSSR limits and values
import pandas as pd
df_sheet1 = pd.read_excel('param_updated_with_all_limits_and_values.xlsx')
# df_master = pd.read_excel('Master.xlsx', sheet_name='Parameters Details')
df_master=df_new_Master

# Function to get unit from master sheet
def get_unit_from_master(commodity, parameter):
    # Find matching row in master sheet
    master_row = df_master[(df_master['Commodity'] == commodity) & 
                          (df_master['Parameter'] == parameter)]
    
    if not master_row.empty:
        unit = master_row['Units'].iloc[0]
        # Handle NaN or empty units
        if pd.isna(unit) or str(unit).strip() == '':
            return None
        return str(unit).strip()
    return None

# Iterate through each row in Sheet1
for idx, row in df_sheet1.iterrows():
    commodity = row['Commodity']
    
    # Get all FSSR Limits columns
    fssr_columns = [col for col in df_sheet1.columns if col.startswith('FSSR Limits')]
    
    for fssr_col in fssr_columns:
        # Extract the parameter name by removing 'FSSR Limits ' prefix
        param = fssr_col.replace('FSSR Limits ', '')
        
        # Get unit from master sheet
        unit = get_unit_from_master(commodity, param)
        
        if unit is not None:
            current_value = str(df_sheet1.at[idx, fssr_col])
            # Only add unit if it's not already present in the value    
            if unit not in current_value:
                df_sheet1.at[idx, fssr_col] = f"{current_value} {unit}"

# Save the updated DataFrame
df_sheet1.to_excel('param_updated_with_limits_values_and_units.xlsx', index=False)
print("Successfully added units to FSSR Limits columns")
```

```python
# Define the output file
output_file = 'param_updated_with_compliance_Ak_Final.xlsx'

# Create an Excel writer object
with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    # Save the original dataframe as a separate sheet
    ndf.to_excel(writer, sheet_name='Original_Data', index=False)
    
    # Always keep the first 19 columns
    first_19_columns = df_new.columns[:19].tolist()

    # Loop through each unique commodity and create a separate sheet
    for commodity in ndf['Commodity'].unique():
        # Get the parameters to keep for the current commodity
        columnsTokeep = df_new_Master.loc[
            (df_new_Master['Key_Parameter'] == 'No') & 
            (df_new_Master['Commodity'] == commodity), 
            'Parameter'
        ].drop_duplicates().tolist()
        
        # Define the final list of columns to keep
        filtered_columns = first_19_columns.copy()  # Start with the first 19 columns
        for param in columnsTokeep:
            if param in param_columns:
                filtered_columns.extend([
                    param, 
                    f'FSSR Limits {param}', 
                    f'{param} Compliance'
                ])

        # Keep only the selected columns
        filtered_columns = list(dict.fromkeys(filtered_columns))  # Remove duplicates
        filtered_columns = [col for col in filtered_columns if col in ndf.columns]  # Ensure valid columns

        # Filter the dataframe for the current commodity
        commodity_df = ndf[ndf['Commodity'] == commodity][filtered_columns]

        # Save to Excel with a limited sheet name length
        commodity_df.to_excel(writer, sheet_name=commodity[:31], index=False)

print("Successfully saved separate sheets for each Commodity with filtered columns along with the original data.")
```
